```python
import pandas as pd
```

```python
# Ajustar la ruta del archivo
file_path = './TFM_UNIR/data/raw/BINANCE_BTCUSD.P, 60min.csv'
df = pd.read_csv(file_path)
```

```python
# Mostrar las primeras filas del DataFrame
print(df.head())
```

```python
#Explorar los datos
# Mostrar las primeras filas del DataFrame
df.head()

# Mostrar información general sobre el DataFrame
df.info()

# Describir las estadísticas básicas de las columnas numéricas
df.describe()
```

```python
# Eliminar columnas con cualquier valor nulo
df_cleaned = df.dropna(axis=1, how='any')

# Mostrar las primeras filas del DataFrame después de la limpieza
print("\nPrimeras filas del DataFrame después de la limpieza:")
print(df_cleaned.head())

# Mostrar información general sobre el DataFrame después de la limpieza
print("\nInformación general del DataFrame después de la limpieza:")
df_cleaned.info()
```

```python
# Seleccionar solo las columnas especificadas
columns_to_keep = ['time', 'open', 'high', 'low', 'close', 'Bullish.1', 'Bullish+.1', 'Bearish.1', 'Bearish+.1']
df_cleaned = df[columns_to_keep]

# Mostrar las primeras filas del DataFrame filtrado
print(df_cleaned.head())

# Mostrar información general sobre el DataFrame filtrado
df_cleaned.info()
```

```python
# Guardar los datos procesados
df_cleaned.to_csv('./TFM_UNIR/data/processed/BINANCE_BTCUSD_cleaned.csv', index=False)
```
